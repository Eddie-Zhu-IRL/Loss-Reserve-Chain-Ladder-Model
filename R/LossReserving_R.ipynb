## Import Libraries: tidyverse and ChainLadder

```r
library(tidyverse)
suppressPackageStartupMessages(library(ChainLadder))
library(openxlsx)
```

## Data Processing

### 1. Read Data

```r
data <- read_csv("ppauto_pos98-07.csv", show_col_types = FALSE)
head(data)
```

### 2. User Input for a company

```r
# ask user to type in an integer as company_id to locate a company
# id can be found in Excel sheet
company_id <- as.integer(readline(prompt = "Enter your name: "))
print(company_id)
```

```r
# create a list of companies
(company_list <- data %>% select ("GRNAME") %>% distinct() %>% pull())
```

```r
# print user's selected company for further reconciliation
print(paste("User input: ", company_list[company_id], sep = ""))
```

## Create Triangle Loss Data

### 1. Create Loss Triangle Data using triangle function in package ChainLadder

```r
triangle_data <- data %>% 
                        filter(GRNAME == company_list[company_id], DevelopmentYear < 2008) %>% 
                        as.triangle( origin = "AccidentYear", dev = "DevelopmentLag", value = "IncurredLosses")
```

```r
# view triangle data
triangle_data
```

### 2. Create Loss Triangle Data Manually

```r
# create triangle data by pivoting for validation
data %>% 
    filter(GRNAME == company_list[company_id], DevelopmentYear < 2008) %>% 
    select(AccidentYear, DevelopmentLag, IncurredLosses) %>% 
    pivot_wider(names_from = DevelopmentLag, values_from = IncurredLosses)
```

## Chain Ladder Modelling: using Mack's method

```r
mack.cl <- MackChainLadder(triangle_data)
summary(mack.cl)
```

### Ultimate Loss and IBNR

```r
(output <- summary(mack.cl)$ByOrigin %>% rownames_to_column("AccidentYear"))
```

## Write Ouput to Excel Spreadsheet for Reconciliation

```r
write.xlsx(output, file = "R_Output.xlsx", sheetName = "R")
```
